```python
def artrans(x, phi_tr, lag_max = 25, plottr = "TRUE"):

    p_tr = len(phi_tr)
    n = len(x)
    nmp = n - p_tr
    nmp1 = nmp - 1

    if (lag_max > nmp1):
        lag_max = nmp1

    aut = acf(x, lag_max, plot = False)$acf
    lag_max1 = lag_max + 1
    k = 0:lag_max
    aut_orig = aut[1:lag_max1]
    y = rep(0, nmp)

    for t in (p_tr:n):
        y[t - p_tr] = x[t]
        for i in (1:p_tr):
            y[t - p_tr] = y[t - p_tr] - phi_tr[i] * x[t - i]

    aut = acf(y, lag_max, plot = False)$acf
    aut.tr = aut[1:lag_max1]
    
    if (plottr == TRUE) {
        numrows <- 2
        numcols <- 2
        fig.width <- 5.5
        fig.height <- 4.5
        cex.labs <- c(0.8, 0.7, 0.9)
        par(mfrow = c(numrows, numcols), mar = c(3.8, 2.5, 1, 
            1))
        tt = 1:n
        plot(tt, x, type = "o", xaxt = "n", yaxt = "n", cex = 0.5, 
            pch = 16, cex.lab = 0.75, cex.axis = 0.75, lwd = 0.75, 
            xlab = "", ylab = "")
        axis(side = 1, cex.axis = 0.9, mgp = c(3, 0.15, 0), tcl = -0.3)
        axis(side = 2, las = 1, cex.axis = 0.9, mgp = c(3, 0.4, 
            0), tcl = -0.3)
        mtext(side = c(1, 2, 1), cex = cex.labs, text = c("Time", 
            "", "Realization"), line = c(1, 1.1, 2.1))
        plot(k, aut_orig, type = "h", xaxt = "n", yaxt = "n", 
            cex = 0, cex.lab = 0.75, cex.axis = 0.75, lwd = 0.75, 
            xlab = "", ylab = "", ylim = c(-1, 1))
        abline(h = 0)
        axis(side = 1, cex.axis = 0.8, mgp = c(3, 0.15, 0), tcl = -0.3)
        axis(side = 2, las = 1, cex.axis = 0.8, mgp = c(3, 0.4, 
            0), tcl = -0.3)
        mtext(side = c(1, 2, 1), cex = cex.labs, text = c("Lag", 
            "", "Sample Autocorrelations"), line = c(1, 1.1, 
            2.1))
        if (plottr == "TRUE") {
            tt = 1:nmp
            plot(tt, y, type = "o", xaxt = "n", yaxt = "n", cex = 0.5, 
                pch = 16, cex.lab = 0.75, cex.axis = 0.75, lwd = 0.75, 
                xlab = "", ylab = "")
            axis(side = 1, cex.axis = 0.9, mgp = c(3, 0.15, 0), 
                tcl = -0.3)
            axis(side = 2, las = 1, cex.axis = 0.9, mgp = c(3, 
                0.4, 0), tcl = -0.3)
            mtext(side = c(1, 2, 1), cex = cex.labs, text = c("Time", 
                "", "Transformed Realization"), line = c(1, 1.1, 
                2.1))
            k = 0:lag_max
            plot(k, aut.tr, type = "h", xaxt = "n", yaxt = "n", 
                cex = 0, cex.lab = 0.75, cex.axis = 0.75, lwd = 0.75, 
                xlab = "", ylab = "", ylim = c(-1, 1))
            abline(h = 0)
            axis(side = 1, cex.axis = 0.8, mgp = c(3, 0.15, 0), 
                tcl = -0.3)
            axis(side = 2, las = 1, cex.axis = 0.8, mgp = c(3, 
                0.4, 0), tcl = -0.3)
            mtext(side = c(1, 2, 1), cex = cex.labs, text = c("Lag", 
                "", "Sample Autocorrelations"), line = c(1, 1.1, 
                2.1))
        }
        return(y)
    }
```
