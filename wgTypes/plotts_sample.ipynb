```python
function (x, lag.max = 25, trunc = 0, arlimits = FALSE) 
{
    fig.width <- 5.5
    fig.height <- 4.5
    cex.labs <- c(0.8, 0.7, 0.9)
    numrows <- 2
    numcols <- 2
    par(mfrow = c(numrows, numcols), mar = c(3.8, 2.5, 1, 1))
    n = length(x)
    nm1 = n - 1
    if (lag.max > nm1) {
        lag.max = nm1
    }
    t = 1:n
    naut = lag.max * 1
    naut1 = naut + 1
    k = 0:naut
    aut = acf(x, lag.max = n - 1, plot = FALSE)$acf
    if (n < 200) {
        plot(t, x, type = "o", xaxt = "n", yaxt = "n", cex = 0.5, 
            pch = 16, cex.lab = 0.75, cex.axis = 0.75, lwd = 0.75, 
            xlab = "", ylab = "")
    }
    if (n >= 200) {
        plot(t, x, type = "l", xaxt = "n", yaxt = "n", cex = 0.5, 
            pch = 16, cex.lab = 0.75, cex.axis = 0.75, lwd = 0.75, 
            xlab = "", ylab = "")
    }
    axis(side = 1, cex.axis = 0.9, mgp = c(3, 0.15, 0), tcl = -0.3)
    axis(side = 2, las = 1, cex.axis = 0.9, mgp = c(3, 0.4, 0), 
        tcl = -0.3)
    mtext(side = c(1, 2, 1), cex = cex.labs, text = c("Time", 
        "", "Realization"), line = c(1, 1.1, 2.1))
    ul = 2/sqrt(n)
    ll = -ul
    autplt = aut[1:naut1]
    plot(k, autplt, type = "h", xaxt = "n", yaxt = "n", cex = 0, 
        cex.lab = 0.75, cex.axis = 0.75, lwd = 0.75, xlab = "", 
        ylab = "", ylim = c(-1, 1))
    abline(h = 0)
    if (arlimits == TRUE) {
        abline(h = ul, lty = 2)
        abline(h = -ul, lty = 2)
    }
    axis(side = 1, cex.axis = 0.8, mgp = c(3, 0.15, 0), tcl = -0.3)
    axis(side = 2, las = 1, cex.axis = 0.8, mgp = c(3, 0.4, 0), 
        tcl = -0.3)
    mtext(side = c(1, 2, 1), cex = cex.labs, text = c("Lag", 
        "", "Sample Autocorrelations"), line = c(1, 1.1, 2.1))
    freq = (1:floor(n/2))/n
    nf = length(freq)
    pgram = rep(0, nf)
    for (i in 1:nf) {
        cosvector = cos(2 * pi * freq[i] * (1:(n - 1)))
        pgram[i] = aut[1] + 2 * sum(aut[-1] * cosvector)
    }
    db = 10 * log10(pgram)
    nf = length(db)
    min.per = min(db[1:nf])
    if (n <= 200) {
        plot(freq, db, type = "n", xaxt = "n", yaxt = "n", cex = 0.5, 
            pch = 16, cex.lab = 0.75, cex.axis = 0.75, lwd = 0.75, 
            xlab = "", ylab = "")
    }
    if (n > 200) {
        plot(freq, db, type = "l", xaxt = "n", yaxt = "n", cex = 0.5, 
            pch = 16, cex.lab = 0.75, cex.axis = 0.75, lwd = 0.75, 
            xlab = "", ylab = "")
    }
    axis(side = 1, cex.axis = 0.8, mgp = c(3, 0.15, 0), tcl = -0.3)
    axis(side = 2, las = 1, cex.axis = 0.8, mgp = c(3, 0.4, 0), 
        tcl = -0.3)
    mtext(side = c(1, 2, 1), cex = cex.labs, text = c("Frequency", 
        "dB", "Periodogram"), line = c(1, 1.1, 2.1))
    for (i in 1:nf) {
        segments(freq[i], min.per, freq[i], db[i])
    }
    if (trunc == 0) {
        M = floor(2 * sqrt(n))
    }
    else {
        M = trunc
    }
    pzgram = rep(0, nf)
    for (i in (1:nf)) {
        cosvector = cos(2 * pi * freq[i] * (1:(n - 1)))
        weight = c(c(1 - 6 * ((0:floor(M/2))/M)^2 + 6 * ((0:floor(M/2))/M)^3), 
            c(2 * (1 - ((floor(M/2) + 1):M)/M)^3), rep(0, n - 
                M - 1))
        pzgram[i] = aut[1] * weight[1] + 2 * sum(aut[-1] * weight[-1] * 
            cosvector)
    }
    dbz = 10 * log10(pzgram)
    plot(freq, dbz, type = "l", xaxt = "n", yaxt = "n", cex = 0.5, 
        pch = 16, cex.lab = 0.75, cex.axis = 0.75, lwd = 0.75, 
        xlab = "", ylab = "")
    axis(side = 1, cex.axis = 0.8, mgp = c(3, 0.15, 0), tcl = -0.3)
    axis(side = 2, las = 1, cex.axis = 0.8, mgp = c(3, 0.4, 0), 
        tcl = -0.3)
    mtext(side = c(1, 2, 1), cex = cex.labs, text = c("Frequency", 
        "dB", "Parzen Window"), line = c(1, 1.1, 2.1))
    out1 = list(autplt = autplt, freq = freq, db = db, dbz = dbz)
    return(out1)
}
```
